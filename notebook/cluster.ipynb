## 1. Initial Visualization:

```python
# Defining a custom color map for clusters using ListedColormap
cluster_colors = colors.ListedColormap(['#0f0e13', '#112f37', '#438f9d', '#075758', '#0c716b', '#89bf9f', '#d2c9c0'])

# Displaying the custom color map
cluster_colors
```

```python
# Reading a CSV file into a pandas DataFrame
df = pd.read_csv('data/raw/cc.csv')  # Reading the CSV file 'cc.csv' and storing it in the DataFrame 'df'

# Displaying the first 10 rows of the DataFrame
df.head(10)  # Showing the first 10 rows of the DataFrame 'df'
```

```python
# Checking for missing values in the DataFrame and printing the sum of null values for each column
print(df.isnull().sum())

# Creating a heatmap to visualize missing values in the DataFrame
plt.figure(figsize=(14, 7))  # Setting the size of the heatmap figure
sns.heatmap(df.isnull(), cmap=['#89bf9f', '#112f37'])  # Plotting the heatmap with custom color mapping
plt.show()  # Displaying the heatmap
```

```python
# List of variables for analysis
variables = ['MINIMUM_PAYMENTS', 'CREDIT_LIMIT']

# Subplot configuration
plt.figure(figsize=(12, 6))
num_variables = len(variables)

# Define a custom color palette
custom_palette = sns.color_palette(['#89bf9f', '#112f37'])

for i, variable in enumerate(variables):
    # Create subplot
    plt.subplot(1, num_variables, i+1)
    
    # Plot boxplot with custom color palette
    sns.boxplot(x=df[variable], palette=custom_palette)
    
    # Set subplot title
    plt.title(f'Boxplot of {variable}')

    # Outlier calculations
    Q1 = df[variable].quantile(0.25)
    Q3 = df[variable].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[variable] < lower_bound) | (df[variable] > upper_bound)]
    percentage_of_outliers = (len(outliers) / len(df)) * 100

    # Display the percentage of outliers
    print(f'Percentage of outliers in {variable}: {percentage_of_outliers:.2f}%')

# Adjust layout and display the plots
plt.tight_layout()
plt.show()

```

```python
# Displaying summary statistics for specific columns in the DataFrame
df[["CREDIT_LIMIT", "MINIMUM_PAYMENTS"]].describe()
```

## 2. Normality Test (Shapiro-Wilk):

```python
# Defining the number of rows and columns for subplots
num_rows, num_cols = 5, 4

# Creating subplots for each column in the DataFrame (excluding 'CUST_ID')
fig, axs = plt.subplots(num_rows, num_cols, figsize=(12, 12))
axs = axs.ravel()

# Selecting columns for plotting
columns_to_plot = df.columns.drop("CUST_ID")[:num_rows*num_cols]

# Looping through selected columns and creating distribution plots
for i, column in enumerate(columns_to_plot):
    sns.distplot(df[column], 
                 kde_kws={'color':'#787775'}, 
                 hist_kws={'color':'#B9A08D'}, 
                 ax=axs[i])
        
    # Setting titles and formatting for each subplot
    axs[i].set_title(column, color='#545a4c', fontsize=10, y=1.03)
    axs[i].xaxis.grid(False)
    axs[i].yaxis.grid(False)
    axs[i].set(xlabel="")

# Removing extra subplots beyond the selected columns
for i in range(len(columns_to_plot), num_rows * num_cols):
    fig.delaxes(axs[i])

# Adjusting layout and displaying the subplots
plt.tight_layout()
plt.show()
```

```python
# Significance level for the test
alpha = 0.05

# Looping through columns and performing the Shapiro-Wilk test
for column in df.columns:
    if column != "CUST_ID":
        statistic, p_value = stats.shapiro(df[column])
        if p_value > alpha:
            print(f'The column "{column}" follows a normal distribution (does not reject the null hypothesis)')
        else:
            print(f'The column "{column}" does not follow a normal distribution (rejects the null hypothesis)')
```

## 3. Data Standardization:

```python
# Scaling the features in the DataFrame (excluding 'CUST_ID') using StandardScaler
df_scaled = pd.DataFrame(StandardScaler().fit_transform(df.drop(["CUST_ID"], axis=1)))
```

## 4. Hopkins Test:

```python
# Calculating the Hopkins statistic for the scaled DataFrame
hopkins_value = hopkins(df_scaled)

# Printing the result of the Hopkins statistic test
hopkins_result = 'Result: {:.2f}'.format(hopkins_value)
print(hopkins_result)

# Checking if the dataset passed the Hopkins test and is clusterable
if 0.7 < hopkins_value < 0.99:
    print("The dataset passed the Hopkins test, and it is clusterable.")
else:
    print("The dataset did not pass the Hopkins test, and it is not clusterable.")
```

## 5. Dimensionality Reduction (PCA):

```python
# Converting the scaled DataFrame to a NumPy array
df_pca = np.asarray(df_scaled)

# Applying PCA with 2 components
pca = PCA(n_components=2, random_state=0)
df_pca = pca.fit_transform(df_scaled)
```

## 6. Determining the Number of Clusters (Elbow Method):

```python
# Calling the elbow function with the PCA-transformed DataFrame
elbow(df_pca)
```

## 7. Application of K-Means:

```python
# Assigning the PCA-transformed DataFrame to X
X = df_pca

# Creating a KMeans object with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=0, max_iter=500)

# Performing clustering and obtaining cluster labels
y_kmeans = kmeans.fit_predict(X)
```

## 8. Results Visualization:

```python
colors_c = ['#0f0e13', '#438f9d', '#89bf9f' , '#d2c9c0']

# Call the visualizerScatter function with the PCA-transformed DataFrame and KMeans results
visualizerScatter(X, kmeans, y_kmeans, num_clusters=4, colors_c=colors_c)
```

## 9. Detailed Cluster Analysis

```python
# Get the unique values in the 'y_kmeans' array
unique_values = np.unique(y_kmeans)
```

```python
# Increment each element in 'y_kmeans' by 1
y_kmeans = [x + 1 for x in y_kmeans]
```

```python
# Add a new column named "Clusters" to the DataFrame df
df["Clusters"] = y_kmeans
```

```python
# Initialize an empty list to store the count of data points in each cluster
total = []

# Iterate over clusters (assumed to range from 1 to 4)
for i in range(1, 5):
    # Count the number of data points in the current cluster and append to the list
    c = df[df["Clusters"] == i].shape[0]
    total.append(c)

# Calculate the total count of data points across all clusters
total_count = sum(total)

# Iterate over clusters and print the percentage of total data points in each cluster
for i in range(0, 4):
    percentage = total[i] / total_count
    print(f"Cluster {i+1} represents {percentage:.4%} of the total")
```

## 10. Preparation for Feature Analysis:

```python
# Create a new DataFrame df_x by dropping specified columns
df_x = df.drop(["CUST_ID", "Clusters", "TENURE"], axis=1)
```

```python
# Apply Min-Max scaling to each column in the DataFrame
from sklearn.preprocessing import MinMaxScaler

for col in df_x.columns:
    df_x[col] = MinMaxScaler().fit_transform(df_x[[col]])
```

```python
# Assign the 'TENURE' and 'Clusters' columns from the original DataFrame to df_x
df_x[["TENURE", "Clusters"]] = df[["TENURE", "Clusters"]]
```

```python
# Display descriptive statistics for the modified DataFrame df_x
df_x.describe()
```

## 11. Detailed Feature Visualization by Cluster:

```python
# Define a custom color palette for the clusters
custom_palette = sns.color_palette(['#0f0e13', '#438f9d', '#89bf9f' , '#d2c9c0'])

# Plot histograms for each feature in df_x, separated by clusters
for c in df_x.columns:
    # Create a FacetGrid with one column for each cluster
    grid = sns.FacetGrid(df_x, col='Clusters', hue="Clusters", palette=custom_palette, height=6, aspect=1)
    
    # Map histograms for the feature c
    grid.map(plt.hist, c)
    
    # Set y-axis limits for better visibility
    for ax in grid.axes.flat:
        ax.set(ylim=(0, 10))
    
    # Display the plot
    plt.show()

```
